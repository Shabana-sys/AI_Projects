```python
#This assignment is to predict the efforts required for the estimation of test environment projects
#importing library to read the datasets
import pandas as pd
dataset= pd.read_csv("Effort Estimation.csv")
dataset
```

```python
#Pre-processing the data
#Converting the nominal values to numerical data_One Hot encoding to substitute Country data
dataset=pd.get_dummies(dataset, drop_first=True)
for col in dataset.columns:
    if dataset[col].dtype == 'bool':
        dataset[col] = dataset[col].astype(int)
dataset
```

```python
#Defining the Input and Output variables
independent = dataset[['Applications', 'Environments', 'Total Months', 'Country_ID','Country_MY','Country_SG', 'Country_TH']]
dependent = dataset [['Effort']]
```

```python
#Train_Test_Split
from sklearn.model_selection import train_test_split
input_train, input_test, output_train, output_test = train_test_split (independent, dependent, test_size=0.3, random_state=0)
```

```python
#Creating the model
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(input_train, output_train)
```

```python
#Output of Weight and Bias
weight=regressor.coef_
bias=regressor.intercept_
output_pred=regressor.predict(input_test)
```

```python
#Identifying R2 to validate the accuracy of the model
from sklearn.metrics import r2_score
r2=r2_score(output_test, output_pred)
r2
```

```python
#Saving the model
import pickle
filename="MLR_EX1_Model.sav"
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#Loading the model (Can import the model in a separate notebook as needed)
Loaded_model=pickle.load(open("MLR_EX1_Model.sav",'rb'))
```
